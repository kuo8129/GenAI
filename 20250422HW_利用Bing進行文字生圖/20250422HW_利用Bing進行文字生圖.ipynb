# ***《小熊維尼》・角色再設計***
#### ——以極簡可愛風探索角色全新視覺風格

## 風格：極簡可愛風 (Minimalist Kawaii Style)

* 角色外觀
    * 圓潤呆萌的樣貌，採頭大身體小的比例
    * 平靜溫馴的表情
    * 身體前傾、微向左轉

* 線條構圖
    * 使用簡單的線條描繪輪廓
    * 搭配些微的陰影製造微立體感

* 背景色調
    * 以單一柔和色調為背景
    * 無其他裝飾元素，僅聚焦於角色主體

## 生圖歷程

* 原本想生成簡約版的小熊維尼角色集，但因版權限制無法直接使用角色名稱，生成的圖也與原作差異較大，因此改以角色的**動物原型**進行創作。

* 起初僅輸入**日系簡約、單一柔和色調背景**等prompt，但生成的圖片風格仍不一致，角色間生成的結果落差更大，因此從中挑幾張較滿意的圖請AI分析畫風，再多次調整prompt，才找出**極簡可愛風**最符合預期。

* 原先僅設定角色動作是**坐姿**，但因動作與方向差異仍大，因此後來又指定**身體前傾、微向左轉**，希望能提升一致性。不過Bing在判別左右方向上仍不穩定，故最後是挑選視覺效果較佳的圖片來呈現。
