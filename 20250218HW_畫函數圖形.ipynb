```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Ë®≠ÂÆöÊï∏Â≠∏ÂáΩÊï∏
* Ë®≠ÂÆöËßíÂ∫¶thetaÂæû0Âà∞2ùúãÂÖ±ÂàáÊàê1000ÂÄãÈªû
* ËÆìÂúñÂΩ¢Áî¢Áîü‰∫îÂÄãÂπ≥ÊªëÁöÑÁ™ÅËµ∑ÔºåÂΩ¢Êàê‰∫îËßíÊòüÊòü
* x, yÂ∞áÊ•µÂ∫ßÊ®ôËΩâÊèõÊàêÁõ¥ËßíÂ∫ßÊ®ô

```python
theta = np.linspace(0, 2*np.pi,1000)
r = 1 + 0.25 * np.sin(5 * theta)
x = r * np.cos(theta)
y = r * np.sin(theta)
```

# Ë®≠ÂÆöÈ°èËâ≤Êº∏Â±§
* ËÆìyÂÄºÊúÄÂ∞èÁöÑÈªû(ÊúÄ‰∏ãÊñπ)Â∞çÊáâÂà∞ÊúÄÊ∑±ÁöÑÈ°èËâ≤ÔºåyÂÄºÊúÄÂ§ßÁöÑÈªû(ÊúÄ‰∏äÊñπ)Â∞çÊáâÂà∞ÊúÄÊ∑∫ÁöÑÈ°èËâ≤
* Ë®≠ÂÆöÈÄèÈÅéÊîπËÆäyÂÄº‰æÜÂΩ±ÈüøautumnÈ°èËâ≤Êº∏Â±§

```python
norm = plt.Normalize(vmin=np.min(y), vmax=np.max(y))
colors = plt.cm.autumn(norm(y))
```

# Áπ™Ë£ΩÂúñË°®
ÂúñÂΩ¢Ë®≠ÁΩÆ
 * Áî®forËø¥ÂúàËÆìÂúñÂΩ¢Áî¢ÁîüÊº∏Â±§È°èËâ≤
 * Êõ¥ÊîπÁ∑öÊ¢ùÁ≤óÁ¥∞ËàáÈÄèÊòéÂ∫¶Ë£ΩÈÄ†Âá∫ÂÖâÊöàÊïàÊûú

ÂúñË°®Ë®≠ÁΩÆ
 * Ë®≠ÂÆöÊ®ôÈ°åÊñáÂ≠óÊ†ºÂºè
 * ‰ΩøxyËª∏ÊØî‰æãÁõ∏Á≠âÔºåÈÅøÂÖçÂúñÂΩ¢ËÆäÂΩ¢
 * Ë®≠ÂÆöÊ†ºÁ∑öÊ®£Âºè

```python
# ÂúñÂΩ¢Ë®≠ÁΩÆ
plt.figure(figsize=(5,5))
for i in range(len(x) - 1):
    plt.plot(x[i:i+2], y[i:i+2], color=colors[i], linewidth=2)
plt.plot(x, y, color='gold', linewidth=6, alpha=0.3)

# ÂúñË°®Ë®≠ÁΩÆ
plt.title("Smooth Star", fontsize=14, fontweight='bold', color='orange', loc='center')
plt.axis('equal')
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
